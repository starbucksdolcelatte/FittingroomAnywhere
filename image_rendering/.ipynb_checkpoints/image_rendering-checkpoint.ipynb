```python
import cv2
import numpy as np

generated = cv2.imread("fakeA_resized.jpg", cv2.IMREAD_COLOR)
background = cv2.imread("background.png", cv2.IMREAD_COLOR)

out = np.where(background==[0,0,0], generated, background).astype(np.uint8)
```

```python
cv2.imshow('out', out)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
